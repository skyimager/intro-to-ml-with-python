### using urllib to read html data from webpages

```python
from urllib.request import urlopen
x = urlopen('https://www.bbc.co.uk/')
k = x.read()
```

```python
k
```

```python
parts=str(k).split("<a href=")
parts
```

```python
parts[1:5]
```

```python
for part in parts[1:5]:
    print(part.split('"')[1])
```

```python
in_links=[]
parts=str(k).split("<a href=")

for part in parts:
    link=part.split('"')[1]
    if "www.bbc" not in link:pass
    else:
        in_links.append(link)
in_links
```

```python
len(in_links)
```

```python
in_links=set(in_links)
len(in_links)
```

### Accessing google search with urllib

```python
url='https://www.google.com/search?q=python+programming+tutorials'
resp = urlopen(url)
respData = resp.read()
print(respData) #Error as google prevents request from code. 
```

```python
from urllib.request import Request
url = 'https://www.google.com/search?q=python+programming+tutorials'

headers = {}
headers['User-Agent'] = r"Mozilla/5.0 (X11; Linux i686) AppleWebKit/537.17 (KHTML, like Gecko) Chrome/24.0.1312.27 Safari/537.17"

req = Request(url, headers = headers)
resp = urlopen(req)
respData = resp.read()
```

### Parsing with BeautifulSoup

```python
# !pip install bs4
```

```python
from bs4 import BeautifulSoup
```

```python
mydata = urlopen("https://www.moneycontrol.com/india/stockpricequote/ironsteel/hindalcoindustries/HI").read()
soup = BeautifulSoup(mydata,'lxml')
soup
```

```python
soup.findAll("a") #extract all tags "a"
```

```python
all_links=soup.findAll("a")
for elem in all_links:
    link=str(elem.get('href'))
    print(link)
```

```python
len(soup.findAll('table'))
```

```python
mytable=soup.findAll('table')[4]
mytable.findAll('tr') #rows
```

```python
for rows in mytable.findAll('tr'):
    data=rows.findAll('td') #table-data
    print_data=[]
    for d in data:
        print_data.append(d.text)
    print(print_data)
```
